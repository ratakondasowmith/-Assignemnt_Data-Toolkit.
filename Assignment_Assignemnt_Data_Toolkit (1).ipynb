2.How does broadcasting work in NumPy?

A.Broadcasting in NumPy is a powerful mechanism that allows arithmetic operations on arrays of different shapes without explicitly replicating data.

What Is Broadcasting?
When performing operations on arrays of different shapes, NumPy automatically stretches or "broadcasts" the smaller array across the larger one so element-wise operations can be performed.

 Broadcasting Rules:
To broadcast two arrays together, NumPy compares their shapes element-wise from right to left (trailing dimensions). Two dimensions are compatible when:

They are equal, or

One of them is 1

If these conditions are not met, broadcasting will result in an error.

-Broadcasting allows for faster computation and less memory usage

-You don’t need to manually reshape or tile arrays

-Must follow shape compatibility rules

6.What are the differences between NumPy arrays and Python lists?

A.Data Type Consistency

NumPy Arrays: Homogeneous. All elements in a NumPy array must be of the same data type, which allows for efficient memory and performance optimizations.

Python Lists: Heterogeneous. Elements can be of different data types, which adds flexibility but reduces performance.

Performance;

NumPy Arrays: Offer better performance for numerical computations due to optimized C backend and vectorized operations.

Python Lists: Slower in numerical processing because operations often require explicit loops written in Python.

Memory Usage;

NumPy Arrays: Memory efficient because they store data in contiguous memory blocks and use fixed-size data types.

Python Lists: Less memory efficient as they store references to objects, requiring more memory for the same amount of data.

Mathematical Operations;

NumPy Arrays: Support element-wise operations (e.g., addition, subtraction) natively and efficiently.

Python Lists: Do not support element-wise operations directly; manual looping or use of map() is required.

Functionality;

NumPy Arrays: Provide a wide range of built-in functions for statistical, linear algebra, and logical operations.

Python Lists: Limited to basic operations like append, remove, sort, etc., with no built-in mathematical capabilities.

Multidimensional Support;

NumPy Arrays: Naturally support multi-dimensional data (e.g., matrices, tensors) using the ndarray object.

Python Lists: Multi-dimensional structures must be created using nested lists, which are harder to manage and less efficient.

Broadcasting;

NumPy Arrays: Allow broadcasting, which enables operations on arrays of different shapes without explicit replication.

Python Lists: Broadcasting is not supported; operations must match in shape or be manually adjusted.

Indexing and Slicing;

NumPy Arrays: Offer powerful and flexible indexing, slicing, and boolean masking.

Python Lists: Support only basic slicing and indexing.

8.What does the term "vectorized operation" mean in NumPy?

A.In NumPy, a vectorized operation refers to performing operations on entire arrays (vectors, matrices, etc.) without writing explicit loops. These operations are internally optimized and executed in compiled C code, making them much faster and more efficient than manual iteration using Python for loops.

Key Features of Vectorized Operations:
High performance: Takes advantage of low-level optimizations and SIMD instructions.

Concise code: Eliminates the need for manual loops.

Readable and cleaner: Expresses complex operations with simple syntax.








```python
#Example:
import numpy as np

# Without vectorization (slow)
a = [1, 2, 3, 4]
b = [2*x for x in a]  # List comprehension

# With vectorization (fast)
arr = np.array([1, 2, 3, 4])
result = 2 * arr
print(result)  # Output: [2 4 6 8]
```

21.Why is NumPy's array slicing faster than Python's list slicing?

A.NumPy's array slicing is faster and more efficient than Python’s list slicing due to how NumPy arrays are stored and accessed in memory, and how operations are implemented under the hood.

Key Reasons:

 1. Contiguous Memory Allocation
NumPy arrays are stored in a single, contiguous block of memory (like C arrays).

 Python lists are arrays of pointers to objects scattered in memory, causing more overhead during slicing.

 2. No Data Copy (Views instead of Copies)
Most NumPy slices return views, not copies, so they don’t duplicate data, making them faster and memory-efficient.

3. Implemented in C
NumPy slicing is handled by optimized C code, avoiding Python-level overhead.

4. Fixed Data Types
NumPy arrays have a homogeneous data type, enabling faster calculations and indexing.

 Python lists can hold mixed types, adding complexity and slowing down access.


```python
#Speed Comparison Example:
import numpy as np
import time

# Python list slicing
py_list = list(range(10_000_000))
start = time.time()
sliced_list = py_list[1_000:2_000]
print("Python list slicing:", time.time() - start)

# NumPy array slicing
np_array = np.arange(10_000_000)
start = time.time()
sliced_array = np_array[1_000:2_000]
print("NumPy array slicing:", time.time() - start)
```
